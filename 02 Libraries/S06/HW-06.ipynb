```python
def f(x):
    return np.sin(1+ np.exp(-1 * x ** 2)) 
```

داده‌های زیر به منظور مطالعه اثر سرعت باد (x) بر میزان تولید برق (y) در توربین بادی جمع آوری شده است. در قالب نمودار پراکندگی رابطه بین این دو را بررسی کنید. سپس خط رگرسیون و منحنی LOWESS را روی نقاط برازش دهید. کدام گزینه نادرست است؟ 
</div>

```python
df = pd.DataFrame({
    'x':[5.0,6.0,3.4,2.7,10.0,9.7,9.55,3.05,8.15,6.20,2.90,6.34,4.60,5.80,7.40,3.60,7.85,8.80,7.0,5.45,9.10,10.20,4.10,3.95,2.45],
    'y':[1.582,1.822,1.057,0.500,2.236,2.386,2.294,0.558,2.166,1.866,0.653,1.930,1.562,1.737,2.088,1.137,2.179,2.112,1.800,1.501,2.303,2.310,1.194,1.440,0.123]
})
```

```python
import statsmodels.api as sm
plt.scatter(df.x,df.y)
params = np.polyfit(df.x,df.y,1)
values = np.linspace(df.x.min(),df.x.max(),100)
reg = np.polyval(params,values)

plt.scatter(6,np.polyval(params,6),color='red')
plt.text(6+0.25,np.polyval(params,6)-0.1,f"{round(np.polyval(params,6),3)}",color='red')

print(params)

plt.plot(values,reg,color='red')
lowess = sm.nonparametric.lowess(df.y,df.x)
plt.plot(lowess[:,0],lowess[:,1],color='orange')

plt.scatter(6,list(lowess[:,1])[(list(lowess[:,0]).index(6))],color='orange')
plt.text(6-1,list(lowess[:,1])[(list(lowess[:,0]).index(6))]+0.1,f"{round(list(lowess[:,1])[(list(lowess[:,0]).index(6))],3)}",color='orange')

plt.grid()
plt.show()
```
