<div dir='rtl'>۴. براساس داده‌ها، برآورد از میانگین این متغیر در جامعه 9.63 است.</div>



```python
data = pd.Series([10.22, 9.35, 11.39, 11.42, 8.76, 7.33, 9.09, 8.64, 9.42,
5.23, 9.82, 11.63, 9.97, 11.2 , 11.27, 7.3 , 11.09, 7.9 , 8.31, 13.35])


plt.hist(data,bins = 10, alpha = 0.7, density = True)
plt.axvline(data.mean(),color='red',
            label=round(data.mean(),2),
            linewidth = 2, 
            linestyle = '--')
plt.title('Histogram of data')
plt.xlabel('Data')
plt.ylabel('Density')

density = stats.gaussian_kde(data)
xp = np.linspace(int(data.min()),
                 int(data.max()),
                 100)
yp = density.pdf(xp)
plt.plot(xp,yp,linewidth=2)

plt.legend()
plt.show()
sm.qqplot(data.values,line='s')
```
